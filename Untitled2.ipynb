```python
import pandas as pd

# Load the dataset from '/content/iris.xls'
try:
  df = pd.read_excel('/content/iris.xls')
  print("Dataset loaded successfully.")
  display(df.head())
except FileNotFoundError:
  print("Error: '/content/iris.xls' not found. Please ensure the file is in the correct location.")
except Exception as e:
  print(f"An error occurred: {e}")
```

```python
print("Columns in the dataset:")
print(df.columns)
```

```python
print("Mean of each column:")
print(df.mean(numeric_only=True))
```

```python
print("Null values in each column:")
print(df.isnull().sum())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

numerical_columns = ['SL', 'SW', 'PL', 'PW']

for col in numerical_columns:
  plt.figure(figsize=(8, 6))
  sns.histplot(data=df, x=col, kde=True)
  plt.title(f"Distribution of {col}")
  plt.xlabel(col)
  plt.ylabel("Frequency")
  plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.pairplot(df, hue='Classification', markers=["o", "s", "D"])
plt.suptitle('Pairwise Scatter Plots of Numerical Features by Classification', y=1.02)
plt.show()
```

```python
sns.countplot(data=df, x='Classification')
plt.title("Distribution of Iris Species")
plt.xlabel("Species")
plt.ylabel("Count")
plt.show()
```
